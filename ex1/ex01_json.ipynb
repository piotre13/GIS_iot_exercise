The city has deployed a network of environmental sensors across its neighbourhoods.
Each sensor sends data as **JSON** — the universal language of IoT data exchange.

In this exercise you will:
- Read and write JSON in Python
- Navigate nested structures
- Filter, transform and summarise sensor data

---

## 1️⃣  Meet the data

Below is a JSON string representing one reading from a sensor in Piazza Castello.
Run the cell and inspect what Python gives you.

```python
import json

raw = '''
{
    "sensor_id": "TRN-001",
    "location": "Piazza Castello",
    "active": true,
    "readings": {
        "temperature_c": 22.4,
        "humidity_pct": 58,
        "pm25_ugm3": 12.1
    },
    "tags": ["air-quality", "weather", "centre"]
}
'''

sensor = json.loads(raw)   # string → Python dict
print(type(sensor))        # should be <class 'dict'>
print(sensor)
```

## 2️⃣  Access nested values

JSON objects become Python **dicts**, arrays become **lists**.

```python
print("Sensor ID  :", sensor["sensor_id"])
print("Temperature:", sensor["readings"]["temperature_c"], "°C")
print("First tag  :", sensor["tags"][0])
print("Is active? :", sensor["active"])
```

## 3️⃣  A network of sensors

Now we have **five sensors** across the city. Let's work with a list of them.

```python
network = [
    {"sensor_id": "TRN-001", "location": "Piazza Castello",  "active": True,  "readings": {"temperature_c": 22.4, "humidity_pct": 58, "pm25_ugm3": 12.1}},
    {"sensor_id": "TRN-002", "location": "Piazza Vittorio",  "active": True,  "readings": {"temperature_c": 23.1, "humidity_pct": 61, "pm25_ugm3": 18.4}},
    {"sensor_id": "TRN-003", "location": "Mirafiori",        "active": False, "readings": {"temperature_c": 21.0, "humidity_pct": 70, "pm25_ugm3": 35.2}},
    {"sensor_id": "TRN-004", "location": "Lingotto",         "active": True,  "readings": {"temperature_c": 24.5, "humidity_pct": 52, "pm25_ugm3": 22.7}},
    {"sensor_id": "TRN-005", "location": "Parco del Po",     "active": True,  "readings": {"temperature_c": 20.8, "humidity_pct": 75, "pm25_ugm3": 8.3}},
]

# How many sensors?
print(f"Total sensors: {len(network)}")
```

## 4️⃣  Filter active sensors

The city dashboard should only show **active** sensors.

```python
active = [s for s in network if s["active"]]
print(f"Active sensors: {len(active)}")
for s in active:
    print(f"  {s['sensor_id']} — {s['location']}")
```

## 5️⃣  Find the dirtiest air

PM2.5 > 20 µg/m³ is considered moderate pollution. Which sensors are above the threshold?

```python
THRESHOLD = 20.0

polluted = [s for s in network if s["readings"]["pm25_ugm3"] > THRESHOLD]
print(f"Sensors above {THRESHOLD} µg/m³:")
for s in polluted:
    pm = s["readings"]["pm25_ugm3"]
    print(f"  ⚠️  {s['location']:20s}  PM2.5 = {pm}")
```

## 6️⃣  Compute the average temperature

```python
temps = [s["readings"]["temperature_c"] for s in network]
avg_temp = sum(temps) / len(temps)
print(f"Average temperature across the city: {avg_temp:.1f} °C")
```

## 7️⃣  Write JSON back to a file

`json.dumps()` converts Python → JSON string. `indent` makes it human-readable.

```python
report = {
    "city": "Turin",
    "active_count": len(active),
    "avg_temperature_c": round(avg_temp, 1),
    "alerts": [s["location"] for s in polluted]
}

with open("city_report.json", "w") as f:
    json.dump(report, f, indent=2)

print(json.dumps(report, indent=2))   # also print to screen
```
